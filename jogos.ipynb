```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
jogos = ["Fortnite",
         "League of Legends",
        "Crossfire",
        "Roblox",
        "Pubg",
        "Minecraft",
        "Lost Ark",
        "Counter-strike",
        "Dota 2",
        "Elden Ring"]
```

```python
milhoes = [12.3,
           8.3,
           8.1,
           4.0,
           3.23,
           1.4,
           1.32,
           1.30,
           1.29,
           0.95]
```

```python
topjogos = pd.Series(milhoes, jogos)
topjogos
```

```python
topjogos.plot()
```

Com esse gráfico podemos analisar a diferença de comunidades em cada um de seus jogos, uns sempre tem mais players que outros, simples, esse gráfico nos mostra isso de uma forma mais visual, para termos uma ideia maior da discrepância entre eles.
